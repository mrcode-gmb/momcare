# Data Cleaning & Preprocessing

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
print(f'Duplicate rows:{data.duplicated().sum()}')
```

```python
# Apply text cleaning to User Input
data['Cleaned_Input'] = data['User Input'].apply(clean_text)
```

**Encode labels**

```python
label_encoder = LabelEncoder()
data['Encoded_Intent'] = label_encoder.fit_transform(data['Intent'])
```

```python
print(data.head())
```

```python
print(data.tail())
```

**Vectorization, Train_Test_Split**

```python
# First define your feature and target columns
X_text = data['Cleaned_Input']  # this is your cleaned text data
y = data['Encoded_Intent']      # this is your encoded target label

# Split raw text data before vectorizing
X_train_text, X_test_text, y_train, y_test = train_test_split(X_text, y, test_size=0.2, stratify=y, random_state=42)

# Fit vectorizer only on training data to avoid data leakage
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)

# Save the vectorizer
joblib.dump(vectorizer, 'tfidf_vectorizer.joblib')
joblib.dump(label_encoder, 'label_encoder.joblib')
```

# Model Training & Evaluation

```python
# Dictionary of models to train and evaluate
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(),
    "Naive Bayes": MultinomialNB(),
    "Random Forest": RandomForestClassifier()
}
```

```python
# Function to train and evaluate model
def train_evaluate_model(model, model_name):
 print(f"\n--- {model_name} ---")
 model.fit(X_train, y_train)
 y_pred = model.predict(X_test)

 accuracy = accuracy_score(y_test, y_pred)
 precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
 recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
 f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)  

 print("Accuracy:", accuracy)
 print("Precision:", precision)
 print("Recall:", recall)
 print("F1 Score:", f1)
 print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))
    
 print("\nClassification Report:\n", classification_report(y_test, y_pred))
    
 cm = confusion_matrix(y_test, y_pred)
 sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
 plt.title(f'Confusion Matrix for {model_name}')
 plt.xlabel('Predicted')
 plt.ylabel('Actual')
 plt.show()
```

```python
# Train and evaluate all models
for name, model in models.items():
    train_evaluate_model(model, name)
```

**Saving the Best Model**

```python
def save_best_model(model, model_name, file_name="best_chatbot_model.joblib"):
    """
    Save the best performing model using joblib.
    
    Parameters:
    - model: The trained ML model to save
    - model_name: A string name for the model (for logging purposes)
    - file_name: File path to save the model (default is 'best_chatbot_model.joblib')
    """
    joblib.dump(model, file_name)
    print(f"✅ {model_name} saved successfully as '{file_name}'")
```
